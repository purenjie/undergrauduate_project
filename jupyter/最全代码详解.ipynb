```python
import pywt
import pandas as pd
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
```

读取 excel 文件，默认返回第一张表（sheet1），返回类型为 DataFrame

```python
# 读取 excel 文件，默认返回第一张表
# sheet:1000 rows × 7 columns
# type(sheet)：<class 'pandas.core.frame.DataFrame'>
def get_data(file):
    sheet = pd.read_excel(io=file)
    return sheet

excel_file = '/home/solejay/program/undergrauduate_project/excel/1000_1.xlsx'
data = get_data(excel_file)
data
```

对 DataFrame 进行切片，x 为输入变量，有 6 个；y 为输出变量（一氧化碳利用率），有 1 个

```python
x = data.iloc[:, 0:6]
y = data.iloc[:, 6] 

print(x)
print(y)
```

```python
对数据进行预处理，将异常值剔除。采取拉依达法则，当数据和平均数偏差大于 3 倍标准差时剔除
```

```python
# 异常值处理：——拉依达法则：数据偏差大于三倍标准差剔除
# x: <class 'pandas.core.frame.DataFrame'>
# y: <class 'pandas.core.series.Series'>
# drop(): 删除数据偏差大于三倍标准差的行
def remove_outlier(x, y):
    mean = np.mean(y)  # 平均值
    std = np.std(y)  # 标准差
    
    lower_limit = mean-3*std  # 最小值
    upper_limit = mean+3*std  # 最大值
    
    for i in range(y.shape[0]):
        if y[i]<lower_limit or y[i]>upper_limit:
            x = x.drop(i)
            y = y.drop(i)
    return x, y

print(len(x))
print(len(y))
x, y = remove_outlier(x, y)
print(len(x))
print(len(y))
```

[sklearn.model_selection.train_test_split 官方文档](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

```python
# 划分训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

[机器学习数据预处理——标准化/归一化方法](https://www.cnblogs.com/bjwu/p/8977141.html)

```python
# 去均值和方差归一化
x_scaler = StandardScaler()
x_train = x_scaler.fit_transform(x_train)
x_test = x_scaler.transform(x_test)
print(x_train)
print(x_test)
```

logspace创建等比数列,默认 `base = 10`,`num` 为等分的数量，数量越多分的越细，也更容易找到最合适的参数

[数据标准化+网格搜索+交叉验证+预测](https://www.cnblogs.com/huadongw/p/6380482.html)

[sklearn.model_selection.GridSearchCV 官方文档](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html#sklearn.model_selection.GridSearchCV)

```python
# 交叉验证选择核函数最佳参数
parameters = {'kernel':['rbf'], 'gamma':np.logspace(-5, 3, num=10, base=2),'C':np.logspace(-2, 3, num=5)}
grid_search = GridSearchCV(SVR(), parameters, cv=10, n_jobs=4, scoring='neg_mean_squared_error')
grid_search.fit(x_train,y_train)
```

```python
# 输出模型预测率 并写入日志文件
def write_log(svr_model, x_test, y_test, excel_file):
    
    y_pred = svr_model.predict(x_test)
    
    mse = mean_squared_error(y_test, y_pred)
    print('均方误差：%.4f' % mse) 
    
    mae = mean_absolute_error(y_test, y_pred)
    print('平均误差：%.4f' % mae)

write_log(grid_search, x_test, y_test, excel_file)
```

```python
# 画出预测值和实际值的图像
def plot_graph(svr_model, x_test, y_test):
    sample = [i for i in range(1, len(y_test)+1)]
    sample = np.reshape(sample, (len(sample), 1))
    
    y_pred = svr_model.predict(x_test)

    plt.plot(sample, y_test, color='black', label='y_test')
    plt.plot(sample, y_pred, color='red', label='y_pred')

    plt.xlabel('sample')
    plt.ylabel('utilization %')
    plt.title('Support Vector Regression')
    plt.legend()
    plt.show()

plot_graph(grid_search, x_test, y_test)
```
